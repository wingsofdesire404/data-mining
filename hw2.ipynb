# 2. 操作题

某家电卖场9月份各家电产品销售数据探索. 为了解家电产品的销售额和各部门的售卖情况，需要对数据进行以下操作:
- (1) 使用分布分析方法，绘制各家电产品的销售额柱形图并进行分析。
- (2) 使用贡献度分析方法，绘制各家电产品的销售额帕累托图并进行分析。
- (3) 使用对比分析方法，绘制各部门2020年9月份的家电产品销售数量折线图并进行分析。

## (1) 使用分布分析方法，绘制各家电产品的销售额柱形图并进行分析。

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'].append('SimHei')  # 使用黑体字显示中文
plt.rcParams['axes.unicode_minus'] = False  # 用于正常显示负号
```

```python
data = pd.read_excel('9月份各家电产品销售额.xlsx')  # 读取数据
```

```python
data.head()
```

```python
# 绘制各家电产品的销售额柱形图
x = data['家电名称']
y = data['销售额']
plt.figure(figsize=(8, 6),dpi=144)  # 设置画布大小

plt.barh(x, y)

plt.xlabel('销售额（元）')  # 设置x轴标题
plt.ylabel('家电名称')  # 设置y轴标题
plt.title('各家电产品的销售额柱形图')  # 设置标题

plt.savefig('各家电产品的销售额柱形图.png')
plt.show()  # 展示图片
```

- 分析：
9月份各家电产品销售额电视机最多，其次是电磁炉；电饭煲最少。

## (2) 使用贡献度分析方法，绘制各家电产品的销售额帕累托图并进行分析。

```python
data = pd.read_excel('9月份各家电产品销售额.xlsx',index_col='家电名称')
data = data['销售额'].copy()
data = data.sort_values(ascending=False)
```

```python
# 绘制各家电产品的销售额帕累托图
plt.figure()
fig = plt.figure(figsize=(20,8))
data.plot(kind='bar')
plt.xlabel('家电名称')
plt.ylabel('销售额（元）')
p = 1.0 * data.cumsum() / data.sum()

p.plot(color='r', secondary_y=True, style='-o', linewidth=1)

# plt.annotate(format(p[6], '.4%'), xy=(6, p[6]), xytext=(6 * 0.9, p[6] * 0.9), 
             #arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=.2'))

plt.ylabel('销售比例')
plt.title('各家电产品的销售额')

plt.savefig('各家电产品的销售额帕累托图.png')

plt.show()
```

## (3) 使用对比分析方法，绘制各部门2020年9月份的家电产品销售数量折线图并进行分析。

```python
data = pd.read_excel('9月份各部门产品销售数量.xlsx')
data.head()
```

```python
plt.figure(figsize=(14, 6))

plt.plot(data['日期'], data['销售部'], color='blue', label='销售部', marker='o')
plt.plot(data['日期'], data['产品部'], color='green', label='产品部', marker='s')
plt.plot(data['日期'], data['市场部'], color='red', label='市场部', marker='x')
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.legend()  # 显示图例
plt.title('各部门2020年9月份的家电产品销售数量')
plt.ylabel('销售数量')
plt.xlabel('日期')
plt.savefig('各部门2020年9月份的家电产品销售数量.png')
plt.show()
```
