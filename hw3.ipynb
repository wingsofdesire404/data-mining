# 实训1 数据清洗

## (0) 读取数据文件, 查看数据集基本信息

```python
import pandas as pd
data = pd.read_csv('USER_INFO.csv', encoding='gbk')
data.info()
```

```python
data.describe()
```

## (1) 调用 `duplicated()` 方法查看数据的重复情况.

```python
dup = data.duplicated()
res = [] 
for i in range(0, len(data)):
    if dup[i] == True:
        res.append(i)
print(res)
```

## (2) 调用 `drop_duplicates()` 方法删除重复值.

```python
data.shape
data.drop_duplicates()
```

## (3) 查看数据缺失值, 将性别缺失的赋值为3, 年龄缺失的赋值为0, 删除含有其它缺失值的记录.

```python
data.isnull()
data['CUST_SEX'].fillna(value=3, inplace=True)
data['CERT_AGE'].fillna(value=0, inplace=True)
data.isnull().sum()
```
